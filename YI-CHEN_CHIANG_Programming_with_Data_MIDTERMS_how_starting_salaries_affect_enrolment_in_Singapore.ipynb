<h2>4.2 NUS Web Scrap 2021 Graduate Salaries of different degree </h2>

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text,'html.parser')
#print('Classes of each table:')
for table in soup.find_all('table'):
    #print(table.get('class'))
    table = soup.find('table', class_='tablepress-id-4985')

#print(table)
NUS_gradSalary = []

rows = table.find_all('tr')
for row in rows:
    cells = row.find_all('td')
    if cells:
        row_data = [cell.text for cell in cells]
        NUS_gradSalary .append(row_data)
        
#print(NUS_gradSalary)
```

<h1>5. Comparing the starting salaries of 4 job fields within NUS, NTU and SMU in 2020</h1>

<h2>5.1 Comparing the starting salaries of 4 job fields in NUS in 2020</h2>

```python
nus2020SalariesComputing = []
nus2020SalariesBusiness = []
nus2020SalariesEngineering = []
nus2020SalariesSocial = []

nus2020SalariesOverall = []

nus2020SalariesSector = ['Computing','Business','Engineering','Social Science']

wordsComputing = ['Computing']
wordsBusiness = ['Business']
wordsEngineering = ['Engineering']
wordsSocial = ['Social']

for index,row in df_2020_nus.iterrows():
    degreeTitle = row['Degree']
    salary = row['Median Gross Monthly Starting Salary ($)']
    tokens = nltk.word_tokenize(degreeTitle)

    for token in tokens:
        #COMPUTING
        if token in wordsComputing:
            nus2020SalariesComputing.append(salary)
        
        #BUSINESS
        if token in wordsBusiness:
            nus2020SalariesBusiness.append(salary)
        
        #ENGINEERING 
        if token in wordsEngineering:
            nus2020SalariesEngineering.append(salary)
        
        #SOCIAL SCIENCE
        if token in wordsSocial:
            nus2020SalariesSocial.append(salary)

#COMPUTING
nus2020SalariesComputing = [s.replace(',', '') for s in nus2020SalariesComputing]
nus2020SalariesComputing = [int(s) for s in nus2020SalariesComputing]
nus2020SalariesOverall.append(round(mean(nus2020SalariesComputing),2))

#BUSINESS
nus2020SalariesBusiness = [s.replace(',', '') for s in nus2020SalariesBusiness]
nus2020SalariesBusiness = [int(s) for s in nus2020SalariesBusiness]
nus2020SalariesOverall.append(round(mean(nus2020SalariesBusiness),2))

#ENGINEERING
nus2020SalariesEngineering = [s.replace(',', '') for s in nus2020SalariesEngineering]
nus2020SalariesEngineering = [int(s) for s in nus2020SalariesEngineering]
nus2020SalariesOverall.append(round(mean(nus2020SalariesEngineering),2))

#SOCIAL SCIENCE
nus2020SalariesSocial = [s.replace(',', '') for s in nus2020SalariesSocial]
nus2020SalariesSocial = [int(s) for s in nus2020SalariesSocial]
nus2020SalariesOverall.append(round(mean(nus2020SalariesSocial),2))

#PLOT GRAPH
fig, ax = plt.subplots()
ax.bar(nus2020SalariesSector, nus2020SalariesOverall)
ax.set_ylabel('Mean Starting Salary(SGD $)')
ax.set_title('Staring salary of different job fields of graduates from NUS 2020')
plt.show()
```

From the data above, graduates of NUS Computing in 2020 earns the highest starting salary amongst other job fields, Business, Engineering and Social Science. Hence, concludes that graduates from NUS Computing in 2020 earns the most followed by Business, Engineering then Social Science.
